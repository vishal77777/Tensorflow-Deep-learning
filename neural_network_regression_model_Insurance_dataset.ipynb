3. fitting a model- letting the model try to find patterns between features and labels

```python
tf.random.set_seed(42)

# creating the model

model=tf.keras.Sequential([
                          tf.keras.layers.Dense(1)
                           
                           
])

# compile the model
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"]) #mean abosolute error


model.fit(tf.expand_dims(X,axis=1),y,epochs=100)
```

#lets create a plotting function

```python
def plot_predictions(train_data=x_train,
                     train_labels=y_train,
                     test_data=x_test,
                     test_labels=y_test,
                     predictions=y_pred):
  """
  plot training data ,test data and compares predictions too the ground truth
  """
  plt.figure(figsize=(10,10))

  plt.scatter(train_data,train_labels,c="b",label="training data")
  plt.scatter(test_data,test_labels,c="g",label="training data")

  #plotting y_pred vs y_test:

  plt.scatter(test_data,predictions,c="r",label="predictions")

  plt.legend();

```

```python
#the parameters are default so need to specify again
plot_predictions()
```

```python
#was bored so.... XD
def subplots(x_train,y_train,x_test,y_test,y_pred):

  fig, ax = plt.subplots(3)
  
# set data with subplots and plot
  ax[0].plot(x_train, y_train)
  ax[1].plot(x_test, y_test)
  ax[2].plot(y_test,y_pred)

  
    
  # set the spacing between subplots
  fig.tight_layout()
  plt.show()


subplots(x_train,y_train,x_test,y_test,y_pred)
```

```python
# calculate absolute error
# we squeeze y_pred becoz it is not in the same shape as y_test and it is not in tensor format so convert as u need to get the output
mae=tf.metrics.mean_absolute_error(y_test,tf.squeeze(y_pred))


s=tf.constant(ev[0])
if round(mae.numpy(),3) == round(s.numpy(),3):
  print("verified error number")

mae
```

```python
# caluclate mean squared error
# we squeeze y_pred becoz it is not in the same shape as y_test and it is not in tensor format so convert as u need to get the output

mse=tf.metrics.mean_squared_error(y_test,tf.squeeze(y_pred))
mse
```

```python
# functions to return the error [mae,mse]:

def error(y_true,y_pred,choice):
  if choice =="mae":
    mae=tf.metrics.mean_absolute_error(y_true,tf.squeeze(y_pred))
    error=mae
  elif choice == "mse":
    mse=tf.metrics.mean_squared_error(y_test,tf.squeeze(y_pred))
    error=mse

  return error.numpy()

```

```python
error(y_test,y_pred,"mse")
error(y_test,y_pred,"mae")
```

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the csv file 

insurance= pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
insurance
```

```python
# One hot encoding - to change categorical variables to int variables type

insurance_one_hot=pd.get_dummies(insurance)

insurance_one_hot.head(6)
```

```python
# Create the labels and features (X and Y):
x= insurance_one_hot.drop("charges",axis=1)
y=insurance_one_hot["charges"]

# view X and Y
x.head()
```

```python
# Splitting the dataset for the training and testing of the model 
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x , y, test_size=0.2 , random_state=42)

```

```python
# creating the model
tf.random.set_seed(42)
# TensorFLow early stooping callback for stopping the loop if no improvement in the model
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=5,mode='loss')

insurance_model= tf.keras.Sequential([
                                  tf.keras.layers.Dense(100,name="input_layer"),
                                  tf.keras.layers.Dense(100,name="input_layer1"), 
                                
                                      
                                        
                                 
                                  tf.keras.layers.Dense(1,name="output_layer")
],name="insurance_dataset")

# compliling the model

insurance_model.compile(loss=[tf.keras.losses.mae,tf.keras.losses.mse],
                        optimizer=tf.keras.optimizers.Adam(),
                                                           metrics=["mae","mse"])
# Fitting the model
history=insurance_model.fit(x_train,y_train,epochs=200,callbacks=[callback])

```

```python
# model summary 
#insurance_model.summary()
```

```python
# Evaluating the model 
insurance_model.evaluate(x_test,y_test)
```

```python
#plotting the loss vs epochs grpah

pd.DataFrame(history.history).plot()
plt.ylabel("loss")
plt.xlabel("epochs")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

insurance=pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")

```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
from sklearn.model_selection import train_test_split

# Create a column transformer
ct=make_column_transformer(
    (MinMaxScaler(),["age","bmi","children"]),  #transform these column values between 0 and 1
    (OneHotEncoder(handle_unknown="ignore"),["sex","smoker","region"])

   )   


x= insurance.drop("charges",axis=1)
y=insurance["charges"]


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

# fit the column transformer to training data
ct.fit(x_train)

# Transform training and test data with normalization
x_train_normal=ct.transform(x_train)
x_test_normal=ct.transform(x_test)

```

```python
# Builing the neural network for the same dataset but with a noramlized value 

tf.random.set_seed(42)
# TensorFLow early stooping callback for stopping the loop if no improvement in the model
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=5,mode='loss')

insurance_model_2= tf.keras.Sequential([
                                  tf.keras.layers.Dense(100,name="input_layer"),
                                  tf.keras.layers.Dense(100,name="input_layer1"), 
                                
                                      
                                        
                                 
                                  tf.keras.layers.Dense(1,name="output_layer")
],name="insurance_dataset")

# compliling the model

insurance_model_2.compile(loss=tf.keras.losses.mae,
                        optimizer=tf.keras.optimizers.Adam(),
                                                           metrics=["mae"])
# Fitting the model
insurance_model_2.fit(x_train_normal,y_train,epochs=200,callbacks=[callback])
```

```python
insurance_model_2.evaluate(x_test_normal,y_test)
```

```python
y_predd_insurance=insurance_model_2.predict(x_test_normal)

y_predd_insurance.squeeze,y_test.shape

y_predd_insurance=tf.squeeze(y_predd_insurance)
plot_predictions(y_test,y_predd_insurance)
```
